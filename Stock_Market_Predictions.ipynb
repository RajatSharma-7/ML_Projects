STOCK MARKET PREDICTION PROJECT --> In these Project we'll create Machine Model that how to predict tomorrow's S&P 500 index price using historical data. We'll also perform  how to avoid common issues that make most stock price models overfit in the real world.
We'll start by downloading S&P 500 prices using a package called yfinance.  Then, we'll clean up the data with pandas, and get it ready for machine learning.  

We'll train a random forest model and make predictions using backtesting.  Then, we'll improve the model by adding predictors.  We'll end with next steps you can use to improve the model on your own.

```python
#This Package calls the Yahoo Finance api to Download daily Stock and Index Prices.
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
# MAX -> Query all the data from the very beginning when index was created
sp500 = sp500.history(period = "max")
```

```python
sp500
```

```python
sp500.index
```

```python
#Cleaning And Visualizing Our Stock Market Data
sp500.plot.line(y="Close",use_index=True)
```

```python
#Deleting Unecessary Columns From Data
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#Setting Up Our Target For Machine Learning
#Will The Stock Price GO UP OR DOWN
sp500["Tomorrow"] = sp500["Close"].shift(-1) #Created Columns "Tomorrow" by panda Shift Method
```

```python
#We have column that is showing Tomorrow Price.
sp500
```

```python
#Target is to predict whether the Tomorrow's price is greater than Today's Price  or not.
sp500["Target"] = (sp500["Tomorrow"]>sp500["Close"]).astype(int)
```

```python
#1 -> Represents when the price went UP
#2 -> Represents when the price went down
sp500
```

```python
#Some Of the old data may not be usefull making future predictions.
#So we gonna delete the data before 1990-01-01.
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
#Data Before 1990-01-01 Has Been Deleted.
sp500
```

```python
# Training an Initial Machine Learning Model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100,min_samples_split=100, random_state =1)
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close","Volume", "Open", "High","Low"]
model.fit(train[predictors], train["Target"])
```

```python
#Measuring how accurate the model is .
from sklearn.metrics import precision_score
preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index = test.index)
```

```python
precision_score(test["Target"],preds)
```

```python
combined = pd.concat([test["Target"], preds],axis=1)
```

```python
combined.plot()
```

**Backtesting System**--> Backtesting in machine learning is a technique used to evaluate the performance and effectiveness of a machine learning model using historical data. It involves testing the model on past data to simulate how it would have performed if it had been used to make predictions during that period.

```python
#Building a Backtesting System
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds],axis=1)
  return combined
```

```python
#We take the 10 years of data (2500 or 250 per year), and predict 11th year data and so on.
def backtest(data, model, predictors, start = 2500, step= 250):
  all_predictions = []

  for i in range(start, data.shape[0], step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train, test, predictors, model)
    all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
#Adding Additional Predictors to our model to see if it increase our accuracy
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ration_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
#IMPROVING OUR MODEL
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds >= .6] = 1
  preds[preds < .6] = 0
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds],axis=1)
  return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
#207 times market stock price will go down.
#43 times market stock price will go up.
#Measuring Accuracy
precision_score(predictions["Target"], predictions["Predictions"])
```
