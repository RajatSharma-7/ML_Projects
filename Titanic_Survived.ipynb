**Titanic_Passengers_Survival**-> Decision tree algorithm is used to solve classification problem in machine learning domain. In this Project we will solve Whether the Passengers in Titanic Survived(1) or not(0) problem using decision tree. By Using Features Like Age, Sex, Fare, Pclass, Survived Etc.

```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop('Survived', axis = 'columns') #Drop The Target Column
inputs
```

```python
target = df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
inputs
```

```python
inputs_n = inputs.drop(['PassengerId','Pclass','Age','Fare','Name','Cabin','SibSp','Parch','Ticket','Sex','Embarked'],axis = 'columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
#Check Whether passenger is Alive or not
# Alive = 1
# Death = 0
#(pclass_n, Age_n, Fare_n )
model.predict([[3,22,1]])
```
